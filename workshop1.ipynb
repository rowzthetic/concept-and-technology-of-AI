```python
def create_checkerboard(size):
    if size % 2 != 0:
      raise ValueError("Size must be an even number")
      matrix = np.zeros((size , size), dtype = int)
      for i in range(size):
        for j in range(size):
          if(i + j) % 2 == 0:
            matrix[i,j] = 1
            return matrix

size = 8
checkerboard = create_checkerboard(size)
print(checkerboard)
```

```python
import numpy as np

A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])


A_inv = np.linalg.inv(A)
result_identity = np.dot(A, A_inv)
print("A * A^-1 =\n", result_identity)

AB = np.dot(A, B)
BA = np.dot(B, A)
print("AB =\n", AB)
print("BA =\n", BA)

AB_T = np.transpose(AB)
BT_AT = np.dot(np.transpose(B), np.transpose(A))
print("(AB)^T =\n", AB_T)
print("B^T A^T =\n", BT_AT)


A = np.array([[2, -3, 1],
              [1, -1, 2],
              [3, 1, -1]])

B = np.array([-1, -3, 9])

A_inv = np.linalg.inv(A)
X = np.dot(A_inv, B)
print("Solution X (x, y, z):", X)

det_A = np.linalg.det(A)
eigenvalues, eigenvectors = np.linalg.eig(A)
norm_A = np.linalg.norm(A)

print("Determinant of A:", det_A)
print("Eigenvalues of A:", eigenvalues)
print("Eigenvectors of A:\n", eigenvectors)
print("Norm of A:", norm_A)
```

```python
import numpy as np
import time

def measure_time(func, *args, **kwargs):
    start_time = time.time()
    result = func(*args, **kwargs)
    elapsed_time = time.time() - start_time
    return result, elapsed_time

#Element-wise Addition

list1 = [i for i in range(1_000_000)]
list2 = [i for i in range(1_000_000)]

def list_addition(a, b):
    return [a[i] + b[i] for i in range(len(a))]

_, time_list_add = measure_time(list_addition, list1, list2)
print(f"Element-wise addition using Python lists took: {time_list_add:.4f} seconds")

# Numpy Arrays
array1 = np.arange(1_000_000)
array2 = np.arange(1_000_000)

_, time_numpy_add = measure_time(np.add, array1, array2)
print(f"Element-wise addition using NumPy arrays took: {time_numpy_add:.4f} seconds")

#Element-wise Multiplication

def list_multiplication(a, b):
    return [a[i] * b[i] for i in range(len(a))]

_, time_list_mul = measure_time(list_multiplication, list1, list2)
print(f"Element-wise multiplication using Python lists took: {time_list_mul:.4f} seconds")

_, time_numpy_mul = measure_time(np.multiply, array1, array2)
print(f"Element-wise multiplication using NumPy arrays took: {time_numpy_mul:.4f} seconds")

#Dot Product

def list_dot_product(a, b):
    return sum(a[i] * b[i] for i in range(len(a)))

_, time_list_dot = measure_time(list_dot_product, list1, list2)
print(f"Dot product using Python lists took: {time_list_dot:.4f} seconds")

_, time_numpy_dot = measure_time(np.dot, array1, array2)
print(f"Dot product using NumPy arrays took: {time_numpy_dot:.4f} seconds")

#Matrix Multiplication

matrix1 = [[i for i in range(1000)] for _ in range(1000)]
matrix2 = [[i for i in range(1000)] for _ in range(1000)]

def list_matrix_multiplication(a, b):
    result = [[sum(a[i][k] * b[k][j] for k in range(len(b))) for j in range(len(b[0]))] for i in range(len(a))]
    return result

_, time_list_matrix_mul = measure_time(list_matrix_multiplication, matrix1, matrix2)
print(f"Matrix multiplication using Python lists took: {time_list_matrix_mul:.4f} seconds")

# Numpy Arrays
matrix1_np = np.array(matrix1)
matrix2_np = np.array(matrix2)

_, time_numpy_matrix_mul = measure_time(np.dot, matrix1_np, matrix2_np)
print(f"Matrix multiplication using NumPy arrays took: {time_numpy_matrix_mul:.4f} seconds")
```
